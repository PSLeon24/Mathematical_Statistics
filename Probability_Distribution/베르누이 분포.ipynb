```python
import numpy as np
from scipy.stats import bernoulli, binom
import matplotlib.pyplot as plt
```

# 1. 베르누이 분포(Bernoulli Distribution)

```python
# Probability of success for the Bernoulli distribution
p = 0.6

# Generate 1000 samples from the Bernoulli distribution
bernoulli_samples = bernoulli.rvs(p, size=1000)

# Calculate the mean and variance of the samples
mean = np.mean(bernoulli_samples)
variance = np.var(bernoulli_samples)

print(f"Mean of Bernoulli distribution samples: {mean:.2f}")
print(f"Variance of Bernoulli distribution samples: {variance:.2f}")

# Plot the Bernoulli distribution sample results
plt.hist(bernoulli_samples, bins=2, edgecolor='black', alpha=0.7, rwidth=0.6)
plt.xticks([0, 1])
plt.xlabel('Value (0 or 1)')
plt.ylabel('Frequency')
plt.title('Bernoulli Distribution Sample Results')
plt.show()
```

# 2. 이항분포(Binomial Distribution)

```python
# Parameters for the Binomial distribution: number of trials n and success probability p
n = 10  # Number of trials
p = 0.6  # Probability of success

# Generate 1000 samples from the Binomial distribution
binomial_samples = binom.rvs(n, p, size=1000)

# Calculate the mean and variance of the samples
mean = np.mean(binomial_samples)
variance = np.var(binomial_samples)

print(f"Mean of Binomial distribution samples: {mean:.2f}")
print(f"Variance of Binomial distribution samples: {variance:.2f}")

# Plot the Binomial distribution sample results
plt.hist(binomial_samples, bins=n+1, edgecolor='black', alpha=0.7, rwidth=0.6)
plt.xticks(range(n+1))
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution Sample Results')
plt.show()
```
