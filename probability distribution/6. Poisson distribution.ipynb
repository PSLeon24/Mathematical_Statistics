```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np

def pmf_poisson_graphing(lamb):
    '''
    주어진 lambda에 대한 확률값을 시각화한다.
    x의 범위는 [0, 3*lambda]이다.
    '''
    xs = np.arange(0, 3*lamb+1)
    ps = poisson.pmf(xs, lamb)

    fig, ax = plt.subplots()
    ax.plot(xs, ps, marker='o')

    plt.show()
```

- xs: 포아송 분포의 확률 질량 함수를 계산할 때 사용하는 x 값들의 배열
- lamb: 포아송 분포의 모수(lambda). 포아송 분포의 평균과 분산이 모두 lambda인 하나의 매개변수

```python
pmf_poisson_graphing(3)

pmf_poisson_graphing(7)
```
